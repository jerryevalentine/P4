```python
import pandas as pd
songs_df = pd.read_csv("resources/spotify_songs.csv")
songs_df.head()
```

```python
songs_df_pp = songs_df.drop(["track_id","track_name",'track_artist',"track_album_id",'track_album_name','playlist_name','playlist_id'],axis=1)
songs_df_pp.head()
```

```python
from datetime import datetime

songs_df_pp['track_album_release_date'] = pd.to_datetime(songs_df_pp['track_album_release_date'],format='mixed')
songs_df_pp['release_month'] = songs_df_pp['track_album_release_date'].dt.month
songs_df_pp.head()
```

```python
songs_df_pp_encoded = pd.get_dummies(songs_df_pp)
songs_df_pp_encoded.head()
```

```python
songs_df_pp_encoded['track_popularity'].value_counts()
```

```python
null_values = songs_df_pp_encoded.isnull().sum()
print(null_values)
```

```python
def classify_hit(score):
    if score >= 90:
        return 1
    else:
        return 0
```

```python
songs_df_pp_encoded['hit'] = songs_df_pp_encoded['track_popularity'].apply(classify_hit)
songs_df_pp_encoded.head()
```

```python
songs_df_pp_encoded['release_day'] = songs_df_pp_encoded['track_album_release_date'].dt.day
songs_df_pp_encoded.head()
```

```python
songs_df_pp_encoded = songs_df_pp_encoded.drop(['track_popularity'], axis=1)
songs_df_pp_encoded.head()
```

```python
songs_df_pp_encoded = songs_df_pp_encoded.drop(['track_album_release_date'], axis=1)
songs_df_pp_encoded.head()
```

```python
songs_df_pp_encoded['release_month'] = songs_df_pp_encoded['release_month'].astype(int)
songs_df_pp_encoded['release_day'] = songs_df_pp_encoded['release_day'].astype(int)
```

```python
songs_df_pp_encoded.dtypes
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

y = songs_df_pp_encoded['hit'].values
X = songs_df_pp_encoded.drop(columns="hit").values

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```
