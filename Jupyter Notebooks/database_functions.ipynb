```python
import sqlite3

# SQLite database path
DATABASE = 'C:/Users/jerry/Programming/LRM/resources/LRM.db'

# Function to create the model_results table
def create_model_results_table():
    conn = sqlite3.connect(DATABASE)
    cursor = conn.cursor()

    # Drop the model_results table if it exists
    cursor.execute('DROP TABLE IF EXISTS model_results')

    # Create the model_results table
    cursor.execute('''
        CREATE TABLE model_results (
            model_results_id INTEGER PRIMARY KEY AUTOINCREMENT,
            model_name VARCHAR(255),
            table_name VARCHAR(255),
            r_squared DECIMAL,
            confusion_matrix VARCHAR(1000)
        )
    ''')

    conn.commit()
    conn.close()

if __name__ == '__main__':
    create_model_results_table()
    print("model_results table created successfully.")
```

```python
import sqlite3
import pandas as pd

def get_metadata(database_name):
    conn = sqlite3.connect(database_name)
    metadata = []

    # Query to get all tables in the database
    tables_query = "SELECT name FROM sqlite_master WHERE type='table';"
    tables_df = pd.read_sql_query(tables_query, conn)

    # Iterate through each table to get columns and constraints
    for table in tables_df['name']:
        # Query to get column info
        columns_query = f'PRAGMA table_info("{table}");'
        columns_df = pd.read_sql_query(columns_query, conn)
        
        # Query to get foreign keys
        foreign_keys_query = f'PRAGMA foreign_key_list("{table}");'
        foreign_keys_df = pd.read_sql_query(foreign_keys_query, conn)
        
        # Get column details
        for _, row in columns_df.iterrows():
            column_name = row['name']
            data_type = row['type']
            key = 'PRIMARY KEY' if row['pk'] else ''
            constraint = ''
            
            # Check if the column has any foreign keys
            foreign_key_info = foreign_keys_df[foreign_keys_df['from'] == column_name]
            if not foreign_key_info.empty:
                for _, fk_row in foreign_key_info.iterrows():
                    constraint += f'FOREIGN KEY({fk_row["from"]}) REFERENCES {fk_row["table"]}({fk_row["to"]}) '

            metadata.append((column_name, table, data_type, key, constraint.strip()))

    # Create DataFrame from the metadata
    metadata_df = pd.DataFrame(metadata, columns=['Column Name', 'Table Name', 'Data Type', 'Key', 'Constraint'])
    conn.close()
    return metadata_df

DATABASE_PATH = r'C:\Users\jerry\Programming\LRM\resources\LRM.db'
metadata_df = get_metadata(DATABASE_PATH)
metadata_df.head(100)
```
