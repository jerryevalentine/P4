```python
def bar_charts(df, cols):#cols is a list of columns
    for col in cols:
        plt.figure(figsize = (12, 4))

        print("\n")
        plt.suptitle(f"{col} distribution")
        df[col].value_counts().plot(kind = "bar")
        plt.xlabel(f"{col}")
        plt.ylabel("count")

        plt.show()
        print("\n")
```

```python
df = IMR
cols = ['Report Year', 'Diagnosis Category', 'Treatment Category', 'Treatment Sub Category', 'Determination, Age Range', 'Patient Gender']
bar_charts(df, cols)
```
