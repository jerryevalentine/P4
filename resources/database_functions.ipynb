```python
import sqlite3
import pandas as pd
from sklearn.preprocessing import StandardScaler
import os

def pp_scale_dummies(database_path, table_name, new_table_name, exclude_columns=[]):
    # Connect to the SQLite database
    conn = sqlite3.connect(database_path)
    
    try:
        # Load the table into a pandas DataFrame
        df = pd.read_sql_query(f"SELECT * FROM {table_name}", conn)
        
        # Exclude specific columns if provided
        columns_to_transform = [col for col in df.columns if col not in exclude_columns]
        
        # Separate numeric and categorical columns
        numeric_cols = df[columns_to_transform].select_dtypes(include=['number']).columns
        categorical_cols = df[columns_to_transform].select_dtypes(include=['object']).columns
        
        # Scale numeric columns
        scaler = StandardScaler()
        df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
        
        # Create dummy variables for categorical columns
        df = pd.get_dummies(df, columns=categorical_cols)
        
        # Save the transformed DataFrame to a new table in the SQLite database
        df.to_sql(new_table_name, conn, if_exists='replace', index=False)
        
        print(f"Table '{new_table_name}' created successfully with scaled numeric columns and dummy variables for categorical columns.")
    
    finally:
        # Close the connection
        conn.close()
```

```python
import sqlite3
import os
import pandas as pd

def map_database_column_numeric_boolean(database_path, table_name, column_name, new_column_name, map_top_percent=None, map_half=False, rescale_numeric_value=None):
    # Connect to the SQLite database
    conn = sqlite3.connect(database_path)
    
    # Load the table into a pandas DataFrame
    df = pd.read_sql_query(f"SELECT * FROM {table_name}", conn)
    
    # Ensure only one of the optional arguments is provided
    options_provided = [map_top_percent is not None, map_half, rescale_numeric_value is not None]
    if sum(options_provided) != 1:
        raise ValueError("Only one of map_top_percent, map_half, or rescale_numeric_value should be provided.")
    
    # Create the new column based on the provided criteria
    if map_top_percent is not None:
        threshold = df[column_name].quantile(1 - map_top_percent / 100)
        df[new_column_name] = (df[column_name] >= threshold).astype(int)
    
    elif map_half:
        median_value = df[column_name].median()
        df[new_column_name] = (df[column_name] > median_value).astype(int)
        if len(df) % 2 != 0:
            median_index = df[column_name].nsmallest(len(df) // 2 + 1).index[-1]
            df.loc[median_index, new_column_name] = 0
    
    elif rescale_numeric_value is not None:
        df[new_column_name] = (df[column_name] >= rescale_numeric_value).astype(int)
    
    # Update the database with the new column
    df.to_sql(table_name, conn, if_exists='replace', index=False)
    
    # Commit changes and close connection
    conn.commit()
    conn.close()

```

```python
import sqlite3
import os

def rename_db_columns(database_path, table_name, columns_to_rename):
    # Connect to the SQLite database
    conn = sqlite3.connect(database_path)
    cursor = conn.cursor()
    
    # Get the existing columns from the table
    cursor.execute(f"PRAGMA table_info({table_name})")
    columns_info = cursor.fetchall()
    
    # Extract the column names and types
    existing_columns = {col[1]: col[2] for col in columns_info}
    
    # Create the new column definitions with renamed columns
    new_columns = [f"{columns_to_rename.get(col, col)} {existing_columns[col]}" for col in existing_columns]
    
    # Create a temporary table with the new column definitions
    cursor.execute(f"CREATE TABLE temp_{table_name} ({', '.join(new_columns)})")
    
    # Insert data from the old table to the new table with renamed columns
    old_columns_str = ", ".join(existing_columns.keys())
    new_columns_str = ", ".join([columns_to_rename.get(col, col) for col in existing_columns])
    cursor.execute(f"INSERT INTO temp_{table_name} ({new_columns_str}) SELECT {old_columns_str} FROM {table_name}")
    
    # Drop the original table
    cursor.execute(f"DROP TABLE {table_name}")
    
    # Rename the temporary table to the original table name
    cursor.execute(f"ALTER TABLE temp_{table_name} RENAME TO {table_name}")
    
    # Commit changes and close connection
    conn.commit()
    conn.close()
```

```python
import sqlite3
import os

def clone_db_table(database_path, old_table_name, new_table_name):
    # Connect to the SQLite database
    conn = sqlite3.connect(database_path)
    cursor = conn.cursor()
    
    # Check if the new table already exists
    cursor.execute(f"SELECT name FROM sqlite_master WHERE type='table' AND name='{new_table_name}'")
    if cursor.fetchone():
        print(f"Error: Table '{new_table_name}' already exists.")
        conn.close()
        return
    
    # Create the new table with the same structure as the old table
    cursor.execute(f"CREATE TABLE {new_table_name} AS SELECT * FROM {old_table_name} WHERE 1=0")
    
    # Copy data from the old table to the new table
    cursor.execute(f"INSERT INTO {new_table_name} SELECT * FROM {old_table_name}")
    
    # Commit changes and close connection
    conn.commit()
    conn.close()
    
    print(f"Table '{old_table_name}' has been successfully cloned to '{new_table_name}'")
```

```python
import sqlite3
import os

def delete_db_columns(database_path, table_name, column_names):
    # Connect to the SQLite database
    conn = sqlite3.connect(database_path)
    cursor = conn.cursor()
    
    # Get the existing columns from the table
    cursor.execute(f"PRAGMA table_info({table_name})")
    columns_info = cursor.fetchall()
    
    # Extract the column names
    existing_columns = [col[1] for col in columns_info]
    
    # Create a new column list excluding the columns to be deleted
    new_columns = [col for col in existing_columns if col not in column_names]
    
    if not new_columns:
        print("Error: All columns cannot be deleted.")
        return
    
    # Create a temporary table with the new columns
    cursor.execute(f"CREATE TABLE temp_{table_name} AS SELECT {', '.join(new_columns)} FROM {table_name}")
    
    # Drop the original table
    cursor.execute(f"DROP TABLE {table_name}")
    
    # Rename the temporary table to the original table name
    cursor.execute(f"ALTER TABLE temp_{table_name} RENAME TO {table_name}")
    
    # Commit changes and close connection
    conn.commit()
    conn.close()


```

```python
# Function to execute a SELECT query on a table


# Function to establish a database connection
def get_db_connection(database_path):
    try:
        conn = sqlite3.connect(database_path)
        print("Connected to database.")
        return conn
    except Exception as e:
        print(f"Failed to connect to database: {e}")
        return None

def query_table(table_name, database_path, num_of_rows=None):
    try:
        conn = get_db_connection(database_path)
        if conn is None:
            return pd.DataFrame()
        
        if num_of_rows:
            query = f"SELECT * FROM {table_name} LIMIT {num_of_rows};"
        else:
            query = f"SELECT * FROM {table_name};"

        df = pd.read_sql(query, conn)
        conn.close()
        return df
    except Exception as e:
        print(f"An error occurred querying the database: {e}")
        return pd.DataFrame()
```
